# HEAT MAP VISUALIZATIONS: -
1. The heat map is used to illustrate the correlations between columns of a dataset (only numerical columns).
2. It provides information about strong positive, strong negative, or no correlation.
3. For example,
   * If we consider the correlation between Age and Age, the value is 1, indicating a strong positive correlation.
   * When we look at the correlation between Age and Resting BP, the value is 0.2, indicating a weak correlation.
   * Similarly, the correlation between Age and MaxHR has a value of -0.3, signifying a slight negative correlation.

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True))
```

# JOINT PLOT VISUALIZATIONS: -

1. Joint plot visualizations are used to verify the correlation between two columns or variables, represented in a heat map.
2. For example, we can verify the correlation between (Age & MaxHR), (Oldpeak & HeartDiseases), (Age & Age) columns, and so on.

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data=df)
```

```python
sns.jointplot(x = 'Age', y = 'Age', data=df)
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data=df)
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data=df, kind= 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data=df, kind = 'reg')
```

```python
sns.jointplot(x = 'Cholesterol', y = 'FastingBS', data=df, kind = 'reg')
```
